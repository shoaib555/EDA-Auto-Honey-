# Automobile
source:https://www.kaggle.com/toramky/automobile-dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
au=pd.read_csv("Automobile.csv")
au.size
```

```python
au.shape
```

```python
au.info()
```

```python
au.describe().transpose()
```

```python
sns.distplot(au['normalized_losses'])
plt.ylabel("freq");
```

```python
sns.jointplot(x='engine_size',y='horsepower',data=au);
```

```python
sns.pairplot(au[['engine_size','horsepower','normalized_losses']]);
```

```python
sns.stripplot(au['fuel_type'],au['horsepower']);
```

```python
sns.swarmplot(au['fuel_type'],au['horsepower']);
```

```python
sns.boxplot(au['number_of_doors'],au['horsepower']);
```

```python
sns.barplot(au['number_of_doors'],au['horsepower']);
```

```python
sns.countplot(au['number_of_cylinders'],hue=au['number_of_doors']);
```

```python
sns.pointplot(au['fuel_system'],au['horsepower']);
```

```python
plt.figure(figsize=(10,7))
sns.countplot(au['body_style']);
```

```python

sns.factorplot(x="fuel_type",y="horsepower",col="number_of_doors",data=au,kind="bar")
```

```python
sns.catplot(x="fuel_type",y="horsepower",col="number_of_doors",hue='body_style',data=au,kind='bar');
```

```python
sns.catplot(x="fuel_type",y="horsepower",col="number_of_doors",hue='body_style',data=au,kind='swarm');
```
